## Bring in State shapefile and merge with PPP data

```python
state = "cb_2019_us_state_5m.shp"
map2 = gpd.read_file(state)
```

```python
map2.plot()
```

```python
map2
```

```python
# Filter outlying states

oth = ['GU', 'MP', 'AS', 'PR', 'VI', 'AK', 'HI']    
mask = map2['STUSPS'].apply(lambda x: any(item for item in oth if item in x)) 
```

```python
map2 = map2[~mask]
```

```python
map2.plot()
```

```python
map2 = map2.drop(columns= ['STATEFP', 'STATENS', 'AFFGEOID', 'LSAD'])
```

```python
map2 = map2.rename(columns= {'STUSPS': 'State'})
```

```python
map2.head(5)
```

```python
df = pd.read_csv("ppp_clean.csv")
```

```python
df.shape
```

```python
del df['Unnamed: 0']
```

```python
df = df[df['State'] != 'XX']
```

```python
x = df.pivot_table(values= 'LoanRange',
              index= 'DateApproved',
              columns= 'State',
              aggfunc= np.sum)
```

```python
x = x.fillna(0)
```

```python
x = x.unstack().reset_index()
```

```python
# Merge datasets

ppp_map = map2.merge(x, on= 'State')
```

```python
ppp_map = ppp_map.rename(columns={0: 'Loan Amt'})
```

```python
ppp_map
```

```python
ppp_map = ppp_map.to_crs({'init': 'epsg:5070' })
ppp_map.plot()
```

```python
#ppp_map.to_csv('ppp_map.csv')
```

## Plot cumulative loans received (by state)

```python
cum_loans_bystate = ppp_map.pivot_table(values= 'Loan Amt',
              columns= 'State',
              aggfunc= np.sum).T
```

```python
cloan = cum_loans_bystate.reset_index()
```

```python
cloan = cloan.rename(columns= {'Loan Amt':'Cumulative Loans'})
```

```python
cloan = map2.merge(cloan, on='State')
```

```python
cloan = cloan.to_crs({'init': 'epsg:5070' })
```

```python
cloan
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig, ax = plt.subplots(1, 1, figsize= (15,10), frameon= False)

divider = make_axes_locatable(ax)                               # Align legend bar to plot
cax = divider.append_axes("right", size="5%", pad=0.1)

cloan.plot(column= 'Cumulative Loans',
          cmap='OrRd',
          ax=ax,
          cax=cax,
          legend= True,
          )


ax.set_title('Cumulative PPP Loans', fontsize=24)
ax.axis('off')
```

```python
fig, ax = plt.subplots(1, 1, figsize= (15,10), frameon= False)

cloan.plot(column= 'Cumulative Loans',
          cmap='OrRd',
          ax=ax,
          
          legend= False,
          )

ax.text(.75, .05, 'Source: US Small Business Administration (SBA)', fontsize= 10)
ax.set_title('Cumulative PPP Loans', fontsize=24)
ax.axis('off')
```
